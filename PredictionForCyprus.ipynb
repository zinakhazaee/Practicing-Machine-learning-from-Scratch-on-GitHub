```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
```

# load data

```python
oecd_bli = pd.read_excel("/content/drive/MyDrive/datasets_sample/oecd.xlsx")
gdp_per_capita = pd.read_excel("/content/drive/MyDrive/datasets_sample/gdp_per_capita.xlsx")
```

```python
oecd_bli
```

```python
gdp_per_capita
```

# Prepare Data


```python
country_stats = pd.merge(oecd_bli, gdp_per_capita, left_on= 'country',right_on= 'GDPpercapita'  , how='inner')
```

```python
country_stats
```

```python
X = np.c_[country_stats["value"]]
y = np.c_[country_stats["Life Satisfaction"]]
```

```python
country_stats.plot(kind='scatter', x="value", y='Life Satisfaction')
plt.show()
```

# Select a linear model

```python
model = sklearn.linear_model.LinearRegression()
```

# Train the model

```python
model
model.fit(X, y)
```

# Make a prediction for Cyprus (GDP per capita)

```python
X_new = [[22587]]
print(model.predict(X_new))
```
